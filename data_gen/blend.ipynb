```python
from blending import *
from pathlib import Path
from PIL import Image
import os
import sys
from matplotlib import pyplot as plt
```

```python
syn_root = Path('/mnt/data-ssd/wa/OneToMany_ToolSynSeg/data_gen/syndata')
src_root = Path('/mnt/data-ssd/wa/OneToMany_ToolSynSeg/data_gen/srcdata')
# syn_root
```

```python
# 1 unique tool per blended image (2 seed each tool)
bg_name_list = os.listdir(src_root / 'bg/aug/bg_1_50_sel') # 35 images
blend_options = ['alpha', 'gaussian', 'laplacian']
tool_name_list = os.listdir(src_root / 'fg/endovis18/aug')
blend_root = str(syn_root / 'Syn-S2-F1')

for blend_type in blend_options:
    print('==> Blend type:', blend_type)
    for tool_name in tool_name_list:
        print(tool_name)
        tool_frame_name = os.listdir(src_root / 'fg/endovis18/source' / tool_name / 'image_2')
        for i in range(20): # bg
            bg = np.array(Image.open(str(src_root / 'bg/aug/bg_1_50_sel' / bg_name_list[i])))
            for tool_frame in tool_frame_name:
                for j in range(20): # fg
                    image_name = str(src_root / 'fg/endovis18/aug' / tool_name / 'image_2_50' / tool_frame[:-4]) + '_' + str(j) + '.png'
                    mask_name = image_name.replace('image_2_50', 'mask_2_50_bin')
                    fg = np.array(Image.open(image_name).convert('RGB'))
                    mk = np.array(Image.open(mask_name))
                    bl = blend(fg, mk, bg, 512, 640, blend_type)
                    # resize to 224*224
                    mk = Image.fromarray(mk)
                    mk = mk.resize((224,224), Image.NEAREST)
                    bl = Image.fromarray(bl).resize((224,224), Image.ANTIALIAS)
                    # save
                    name = blend_type+'-'+tool_name[0]+'-'+image_name.split('/')[-1][:-4]+'-bg_'+str(i)+'.png'
                    # print(name)
                    # sys.exit()
                    mk.save(os.path.join(blend_root, blend_type, 'annotations', 'binary', name))
                    bl.save(os.path.join(blend_root, blend_type, 'images', name))
```

```python
# 2 unique tool per blended image (2 seed per tool)
bg_name_list = os.listdir(src_root / 'bg/aug/bg_1_50_sel') # 35 images
blend_options = ['alpha', 'gaussian', 'laplacian']
tool_name_list = os.listdir(src_root / 'fg/endovis18/aug')
blend_root = str(syn_root / 'Syn-S2-F2')

for tool_name in tool_name_list:
    print(tool_name)
    tool_name_cand = [tool_ for tool_ in tool_name_list if tool_ != tool_name]

    tool_frame_name = os.listdir(src_root / 'fg/endovis18/source' / tool_name / 'image_2')
    for i in range(20): # bg
        bg = np.array(Image.open(str(src_root / 'bg/aug/bg_1_50_sel' / bg_name_list[i])))
        for tool_frame in tool_frame_name:
            for j in range(20): # fg
                image_name = str(src_root / 'fg/endovis18/aug' / tool_name / 'image_2_50' / tool_frame[:-4]) + '_' + str(j) + '.png'
                mask_name = image_name.replace('image_2_50', 'mask_2_50_bin')
                tool_name_sel = np.random.choice(tool_name_cand, 1, replace=False)[0]
                image_name_ = np.random.choice(os.listdir(src_root / 'fg/endovis18/aug' / tool_name_sel / 'image_2_50'), 1, replace=False)[0]
                image_name_ = str(src_root / 'fg/endovis18/aug' / tool_name_sel / 'image_2_50' / image_name_)
                mask_name_ = image_name_.replace('image_2_50', 'mask_2_50_bin')                
                for blend_type in blend_options:
                    # blend first image                    
                    fg = np.array(Image.open(image_name).convert('RGB'))
                    mk = np.array(Image.open(mask_name))
                    bl = blend(fg, mk, bg, 512, 640, blend_type)
                    # blend second image                    
                    fg_ = np.array(Image.open(image_name_).convert('RGB'))
                    mk_ = np.array(Image.open(mask_name_))
                    bl_ = blend(fg_, mk_, bl, 512, 640, blend_type)
                    mk__ = np.logical_or(mk, mk_)
                    # print(image_name_)
                    # sys.exit()
                    # resize to 224*224
                    mk__ = Image.fromarray(mk__)
                    mk__ = mk__.resize((224,224), Image.NEAREST)
                    bl_ = Image.fromarray(bl_).resize((224,224), Image.ANTIALIAS)
                    # save
                    name = blend_type+'-T1_'+tool_name[0]+'-'+image_name.split('/')[-1][:-4]+'-T2_'+tool_name_sel[0]+'-'+image_name_.split('/')[-1][:-4]+'-bg_'+str(i)+'.png'
                    # print(name)
                    # sys.exit()
                    mk__.save(os.path.join(blend_root, blend_type, 'annotations', 'binary', name))
                    bl_.save(os.path.join(blend_root, blend_type, 'images', name))
                # sys.exit()
```
